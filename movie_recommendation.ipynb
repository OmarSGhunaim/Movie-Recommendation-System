```python
import pandas as pd
```

```python
df = pd.read_csv("movies.csv")
df.head()
```

```python
def recommend(movie_name):
    genre = df[df["movie"] == movie_name]["genre"].values
    
    if len(genre) == 0:
        return "Movie not found"
    
    genre = genre[0]
    
    recommendations = df[df["genre"] == genre]["movie"].tolist()
    
    recommendations.remove(movie_name)
    
    return recommendations
```

```python
print(recommend("Avengers"))
```

```python
movie = input("Enter a movie name: ")
print("Recommended:", recommend(movie))
```

If a user selects a movie, similar genre movies are suggested
